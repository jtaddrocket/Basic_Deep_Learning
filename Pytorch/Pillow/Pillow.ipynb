```python
import cv2
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
```

```python
# Tải ảnh
img = Image.open("test.jpg")
```

**Chuyển ảnh màu sang Grayscale**

```python
gray_img = img.convert('L')
gray_img.save('test-l.jpg')

plt.imshow(gray_img, cmap='gray')
plt.show()
```

**Chuyển ảnh màu sang Halftone**

```python
# black and white image
halftone_img = img.convert('1')
halftone_img.save('test-1.png')

plt.imshow(halftone_img, cmap='gray')
plt.show()
```

**Thay đổi độ tương phản (Contrast)**

```python
# Tạo đối tượng Enhancer cho độ tương phản
enhancer = ImageEnhance.Contrast(img)

# Số lượng ảnh muốn hiển thị
num_images = 7

# Thiết lập grid để hiển thị ảnh
fig, axs = plt.subplots(1, num_images, figsize=(20, 5))

# Tăng cường độ tương phản và hiển thị ảnh
for i in range(num_images):
    factor = (i + 1) / 4.0
    new_img = enhancer.enhance(factor)
    axs[i].imshow(new_img)
    axs[i].set_title(f'Contrast {factor:.2f}')
    axs[i].axis('off')

plt.show()
```

**Thay đổi độ sáng (Brightness)**

```python
enhancer = ImageEnhance.Brightness(img)
brightness_img = enhancer.enhance(1.8)
# Darker
# new_img = enhancer.enhance(0.8)
brightness_img.save("brightness-test.jpg")
plt.imshow(brightness_img)
plt.show()
```

**Làm mờ (Gaussian blur)**

```python
from PIL import ImageFilter

blur_img = img.filter(ImageFilter.GaussianBlur(radius=20))
blur_img.save("blur-test.png")

plt.imshow(blur_img)
plt.show()
```
